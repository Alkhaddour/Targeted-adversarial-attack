```python
import torch
import torch.nn as nn
import torch.optim as optim

from torchvision import datasets, transforms
from torch.utils.data import DataLoader

import numpy as np
import matplotlib.pyplot as plt
import matplotlib

%matplotlib inline
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

### Define Network

```python
class Flatten(nn.Module):
    def forward(self, x):
        return x.view(x.shape[0], -1)    

FFN_2 = nn.Sequential(Flatten(), 
                            nn.Linear(784,256), 
                            nn.ReLU(),
                            nn.Linear(256,10)).to(device)
```

### Load data

```python
path = "./data"
model_path = "./models/model_FFN2.pt"
mnist_train = datasets.MNIST(root = path, train=True, download=True, transform=transforms.ToTensor())
mnist_test = datasets.MNIST(root = path, train=False, download=True, transform=transforms.ToTensor())
train_loader = DataLoader(mnist_train, batch_size = 128, shuffle=True)
test_loader = DataLoader(mnist_test, batch_size = 128, shuffle=False)
```

### Train model

```python
def epoch(loader, model, opt=None):
    total_loss, acc = 0.,0.
    for X,y in loader:
        X,y = X.to(device), y.to(device)
        yp = model(X)
        loss = nn.CrossEntropyLoss()(yp,y)
        if opt:
            opt.zero_grad()
            loss.backward()
            opt.step()
        
        acc += (yp.max(dim=1)[1] == y).sum().item()
        total_loss += loss.item() * X.shape[0]
    return acc / len(loader.dataset), total_loss / len(loader.dataset)
```

```python
opt = optim.SGD(FFN_2.parameters(), lr=0.1)
for e in range(10):
    train_acc, train_loss = epoch(train_loader, FFN_2, opt)
    test_acc, test_loss = epoch(test_loader, FFN_2)
    print("Epoch {}: Train Loss: {:.6f}\t Train Accuracy: {:.6f}\t Test Loss: {:.6f}\t Test Accuracy: {:.6f}"
          .format(e, train_loss,train_acc,test_loss,test_acc))
```

### Load examples to run attack on them

```python
for X,y in test_loader:
    X,y = X.to(device), y.to(device)
    break
    
def plot_images(X,y,yp,M,N):
    f,ax = plt.subplots(M,N, sharex=True, sharey=True, figsize=(N,M*1.3))
    for i in range(M):
        for j in range(N):
            ax[i][j].imshow(1-X[i*N+j][0].cpu().numpy(), cmap="gray")
            title = ax[i][j].set_title("Pred: {}".format(yp[i*N+j].max(dim=0)[1]))
            plt.setp(title, color=('g' if yp[i*N+j].max(dim=0)[1] == y[i*N+j] else 'r'))
            ax[i][j].set_axis_off()
    plt.tight_layout()
```

### Define attack

```python
def targeted_attack(model, X, y, eps, alpha,iters,dig):
    delta = torch.zeros_like(X, requires_grad=True)
    for t in range(iters):
        yp = model(X + delta)
        loss = (yp[:,dig] - yp.gather(1,y[:,None])[:,0]).sum()
        loss.backward()
        delta.data = (delta + alpha*delta.grad.detach().sign()).clamp(-eps,eps)
        delta.grad.zero_()
    return delta.detach()
```

### Run attack

```python
delta = targeted_attack(FFN_2, X, y,eps= 0.2, alpha=0.01, iters=40, dig=8)
yp = FFN_2(X + delta)
plot_images(X+delta, y, yp, 3, 6)
```
